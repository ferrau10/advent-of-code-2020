```python
import pandas as pd 
import re
```

# Part 1 

```python
df = pd.read_csv('input2.csv', header=None, sep =':', names = ['rule', 'password'])
```

```python
df 
```

```python
df.rule[0]
```

```python
pattern_number = '[0-9]+'
```

```python
re.findall(pattern_number, df.rule[0])
```

```python
re.findall('[a-zA-Z]',df.rule[0] )
```

```python
def min(x):
    return re.findall(pattern_number, x)[0]

def max(x):
    return re.findall(pattern_number, x)[1]

def get_letter(x):
    return re.findall('[a-zA-Z]', x)[0]


df['max'] = df.rule.apply(max).astype(int)
df['min'] = df.rule.apply(min).astype(int)
df['letter'] = df.rule.apply(get_letter)
```

```python
df 
```

```python
def counter(row):
    counter = row['password'].count(row['letter'])
    return counter 
```

```python
df['count'] = df.apply(lambda row: counter(row), axis=1)
```

```python
df 
```

```python
def inside(row):
    if row['count'] <= row['max'] and row['count'] >= row['min']:
        return True
    return False

df['result'] = df.apply(lambda row: inside(row), axis=1)
```

```python
df[df['result'] == True]
```

```python
def position(row):
    if (row['password'][row['min']] == row['letter']) != (row['password'][row['max']] == row['letter']):
        return True
    return False

df['result2'] = df.apply(lambda row: position(row), axis=1)
```
