```python
x = x.split('\n\n')
```

```python
df = pd.DataFrame(x)
```

```python
def replace(x):
    x = x.replace('\n', ' ')
    return x
df[0] = df[0].apply(replace)
```

```python
df 
```

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
```

```python
def get_group_total(row):
    total = 0
    for i in range(len(alphabet)):
        if row.count(alphabet[i]) != 0:
            total = total + 1
    return total
```

```python
df['group'] = df[0].apply(get_group_total)
```

```python
df.group.sum()
```

```python
def get_people(row):
    if row[-1] == ' ':
        return row.count(' ')
    else:
        return row.count(' ') + 1
```

```python
df['people'] = df[0].apply(get_people)
```

```python
df 
```

```python
df.people[0]
```

```python
def get_group_all(row):
    people = get_people(row)
    total = 0
    
    if people == 1:
        for i in range(len(alphabet)):
            if row.count(alphabet[i]) != 0:
                total = total + 1
    else:
        for i in range(len(alphabet)):
            if row.count(alphabet[i]) == people :
                total = total + 1
    return total 
```

```python
df['group_all'] = df[0].apply(get_group_all)
```
